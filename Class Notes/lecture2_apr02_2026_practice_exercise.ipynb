## Practice Exercise

Calculate the motion of a mass on a spring undergoing simple harmonic motion (ignoring damping and friction). 

The position of a mass on a spring as a function of time is given by:

$$
x(t) = A \cos(\omega t)  
$$
 
Where:

* $A$ is the amplitude  
* $\omega$ is the angular frequency: $\omega=\sqrt{k/m}$
* $k$ is the spring constant
* *m* is the mass

Set:

* $A=1.0$ m
* $k=4.0$ N/m
* $m=1.0$ kg
  

**Instructions**

* Calculate the motion from 0 to 10 seconds in 0.1-second steps using either a ```while``` or ```for``` loop.  
* Store the times and positions in two separate lists. 
* Output the times and positions using a f-string.
* BONUS: Plot the motion vs. time using the ```plot``` function from package ```matplotlib.pyplot```. The syntax of ```plot``` is:

```
plot(xvalues, yvalues)
plot.show()
```


````python
"""Simple Harmonic Motion Program 
    By Acamaro Cutcher on Apr 2, 2026
    Calculate the motion from 0 to 10 seconds in 0.1-second steps using either a while or for loop.
    Store the times and positions in two separates lists.
    Output the times and positions using a f-string. 
    Plot the motion vs. time using the plot function matplotlib. 
    Status: Done!
"""

from math import cos, sqrt
from scipy.constants  import g 
import matplotlib.pyplot as plt

# Constants
A = 10
k = 4
m = 1
w = sqrt(k/m)

# Initial time
t = 0
delta_t = 0.1

# Units
d_units = {'t': 's', 'x':'m'}

# Lists
times_list = []
position_list = []

# position
x = 0

while (t <= 10):
    times_list.append(t)
    x = A*cos(w*t)
    position_list.append(x)
    
    print(f"Time: {t:.2f} {d_units['t']}, position: {x:.2f} {d_units['x']}")
    
    t += delta_t 
    

#Plot data

plt.plot(times_list, position_list)

# Title and Labels
plt.title("Simple Harmonic oscillation")
plt.xlabel("Time")
plt.ylabel("X Position")

# 4. Display the plot
plt.show()
    
    
````
