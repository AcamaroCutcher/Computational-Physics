### Practice Exercise 1 - Creating user-defined function

Write a function to calculate the range of a projectile launched at angle $\theta$ with initial velocity $v_o$. 

$$
x = \frac{2v_o^2}{g}\sin(\theta)\cos(\theta)
$$

Prompt the use for $v_o$ and $\theta$, then pass these to your function. Have your function return the range. 

````python
# Practice Exercise 1

from math import radians, degrees, sin, cos
from scipy.constants import g

# User input

v_initial = float(input("Please input initial Velocity: "))
theta_initial = radians(float(input("Please input initial angle (degrees): ")))

# Range Calculation

x_range = 2 * v_initial**2 * sin(theta_initial) * cos(theta_initial) / g

# Print output

print(f"Initial Velocity: {v_initial:0.2f}, initial theta: {degrees(theta_initial)}, range: {x_range:0.2f}")

````
