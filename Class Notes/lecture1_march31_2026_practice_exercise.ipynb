# 2D motion of ball

Write a program that calculates the range of a ball thrown with initial velocity $v_0$ at angle $\theta$. The program needs to prompt the user for $v_0$ and $\theta$. Output the range in meters. Ignore air resistance.

* Import functions from the math module.
* Import the value of *g* from the submodule scipy.constants (if installed)
* Prompt the user for the initial velocity in m/s and launch angle in degrees.
* Convert these into float variables.
* Convert angle in degrees to radians using math function *radians*.  
* Calculate the range.
* Output a sentence that contains the initial velocity, launch angle, and the range in meters to 2 decimal places.
* Use appropriate names for variables. 
* Comment your code.
