### 2D Motion of a Ball - Derivation of Equations: 

#### Initial Velocity
$$
u_x = u \cdot \cos(\theta)
$$
$$
u_y = u \cdot \sin(\theta)
$$

#### Range

By setting the vertical velocity equation

$$
v_y = u_y - gt 
$$

to zero and solving for time 

$$
t = \frac{u_y}{g}
$$

we find the period the ball takes to reach it's highest point. 

The time required for the ball to return to it's initial start position is the same, therefore to find the total time of flight we just need to double the solution we found before:

$$
t =2 \cdot \frac{u_y}{g}
$$

Finally we can calculate the total horizontal displacement (range) of the ball by multiply the total time of flight with it's constant horizontal velocity giving us: 
$$
D = u_x \cdot t
$$
$$
D = u_x \cdot 2 \cdot \frac{u_y}{g} 
$$
$$
D = \frac{2 u^2}{g} \cos(\theta) \sin(\theta)
$$

```python
"""
2D Motion of Ball
by Acamaro Cutcher, Apr 4, 2025
Calculates the range of ball through at an initial velocity
and initial angle. 
Status: Completed
"""

from math import cos, sin, radians, degrees
from scipy.constants import g

# Launch velocity and degree

v_0 = float(input("Please the balls initial velocity (m/s): ")) 
theta_0 = radians(float(input("Please enter the balls starting angle (degrees): "))) 

# Calculate the range of the ball using the derivation shown above we get: 

D = (2 * v_0**2 / g) * cos(theta_0) * sin(theta_0)

# Display to User:

print(f"Initial velocity {v_0} m/s.")
print(f"Initial angle {theta_0} radians ({degrees(theta_0)} degrees).")
print(f"Range of Ball is: {D} m.")


```
