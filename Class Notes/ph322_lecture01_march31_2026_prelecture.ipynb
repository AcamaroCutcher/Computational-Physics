# Importing functions
So far we have used a number of "built-in" Python functions, i.e. *print()*, *input()*, *float()*, and *int()*. These are functions in the core Python library that we can use "out of the box". Over the years developers have written other Python libraries that contain functions we would like to use. These libraries are called Python packages or modules. To use these we have to first download and install them in a location in our file system that Python knows about. 

You can do this at your system's command line prompt using a package manager like 'pip':
> pip install scipy

Once we have the package installed, we can use functions from the package using the **import** command. The packages we will use most often this term are:

* **numpy** - for numerical arrays and math functions that work with arrays
* **matplotlib** - for plotting and visualization
* **scipy** - for constants and other scientific computing tools
* **math** - contains lots of math functions for scalar variables


Let's go through one example of how to import a function from a package.

Assume we want to find the cosine of the angle theta. Is there a cosine function in the standard Python library? Let's try it.


````python
# calculate the cosine of an angle in radians
theta = 1.32
cos(theta)
````

Oops! The interpretator tells us there is no *cos()* function defined. 

We can see what built-in functions exist by looking at the directory of built-functions using the *dir()* command.

````python
dir(__builtins__)
````

If we scroll through the list we find our *float, input*, and *print* functions, but no *cos* function, nor any other math function. 

Where can we find a *cos* function? In the *math* package (makes sense!) To use it, we can import the math package into our current session using the *import* command:

````python
import math
````

We can now get a directory listing of the functions in the math package like we did above:

````python
dir(math)
````

Great, we see the *cos()* function among others. Let's use it.

````python
cos(theta)
````

What? Python says *cos* is not defined. What gives?

When we import an entire package, we have to tell Python where to look for the function by specifying the package name, a period, then the function name:

````python
math.cos(theta)
````

OK, now Python knows where to find the *cos* function and it works. 
You may think it's a pain to have to first refer to the package name. Is there another way?

Yes, one way is like this:

````python
from math import *
````

````python
cos(theta)
````

The asterisk is the wildcard character. It imports all the functions from the *math* package and allows us to use all the math functions without referring to the *math* package.

This seems like a win-win! 

Not necessarily. The same function name may be contained in more than one package. 
For example, there is also a *cos* function in the *numpy* package that works on arrays, whereas the *cos* function in the *math* package works only on scalars. If we were to import all functions from both package using this method, we may not be able to keep them straight. In general it is better not to use " from <package> import * ".
    
There are a cpl of other options however.  

````python
import math as m

m.cos(theta)
````

Here we use the "import <module> as " command. This allows us to use a shorthand abbreviation to refer to the package name when using the function, in this *m* for *math*.
    
We can also do this:

````python
from math import cos
cos (theta)
````

Here we import only the *cos* function from the *math* package. Doing this, we can use the function directly without referencing the module. 

We can use this syntax to import any number of functions from the same package. 

````python
from math import cos, sin, sqrt, exp
sin(theta)
````

Finally, some package contain sub-libraries which we import using both the package and sub-library names. The most common usage is to import the ```pyplot``` sub-package from the package ```matplotlib```:

````python
import matplotlib.pyplot as plt
plt.plot()
````

 We import the *pyplot* submodule as *plt* and use the *plot()* function to create an empty plot. We'll discuss the *matplotlib* module in detail in Chapter 3.

There are standard ways of importing commonly-used packages, which you should follow.

````python
import numpy as np
import matplotlib.pyplot as plt
import math as m
````
