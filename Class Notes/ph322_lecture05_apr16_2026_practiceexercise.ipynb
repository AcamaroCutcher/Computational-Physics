### Practice Exercise: Animate an object in free-fall

Here is a bit of a challenge problem. 

Later this term we will learn numerical methods to integrate differential equations. As a bit of a preview, here's a simple method to integrate a differential equation, called the *Euler-Cromer method*. Here's how it works.

Suppose we have an object in free-fall. If so, the only force acting on it is gravity, so we can write Newton's Second Law for the object as:

$$
\vec{F}_{net} = m\vec{g} = \frac{d\vec{p}}{dt}= \frac{\Delta \vec{p}}{\Delta t}
$$

where $\vec{p}$ is the momentum of the object.

Since $\Delta\vec{p} = \vec{p}_f - \vec{p}_i$, we can write this as:

$$
\vec{p}_f=\vec{p}_i + m\vec{g}{\Delta t}
$$

We can define the average velocity over time $\Delta t$ as:

$$
\vec{v}_{avg} = \frac{\Delta\vec{x}}{\Delta t},
$$

then

$$
\vec{x}_f = \vec{x}_i + \vec{v}_{avg}\Delta t =  \vec{x}_i + \frac{\vec{p}_{avg}}{m}\Delta t
$$

where $m$ is the mass of the object.

Although we don't know $\vec{p}_{avg}$, the Euler-Cromer method makes the approximation

$$
\vec{p}_{avg} \approx \vec{p}_f
$$


Therefore, we can find the position of the object at every time step using:

$$
\vec{x}_f =  \vec{x}_i + \frac{\vec{p}_{f}}{m}\Delta t
$$
