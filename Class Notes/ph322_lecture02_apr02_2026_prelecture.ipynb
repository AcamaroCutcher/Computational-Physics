### ```if``` statement

There are often times we want to execute code only if a certain condition is met. To do so we use the **if** statement. 

The syntax is:

```python
if <logical_expression>:
    indented block of code to execute if expression is true
elif <a_different_logical_expression>:
    indented block of code to execute if this expression is true
else:
    indented block of code to execute if none of the expressions are true
```

Using an IF-ELIF block in Python allows for the implementation of **conditional logic**, enabling the program to execute different code blocks based on varying conditions. This structure enhances the program's flexibility, enabling it to make decisions dynamically based on the inputs or state of the system. 

In a simple case we might want to compare two numbers and perform different actions depending on the outcome. 

````python
# Prompt user for scores. Convert scores to int so we can numerically compare  
A = int(input("Enter the score for player A: "))
B = int(input("Enter the score for player B: "))


# Compare scores and print out the winning player if one score is higher.  
if A>B:
    print("Player A wins!")
elif B>A:
    print("Player B wins!")
else:
    print("It's a tie!")
 
````

The first two lines prompt the user for input and convert the strings into integers.

Next we come to our **IF-ELIF-ELSE** block of code.

The **IF** statement contains a logical expression (is either TRUE or FALSE), followed by a colon. If the expression is TRUE, the indented block is executed. If it is FALSE it skips this block of code and moves to the **ELIF** statement.

The **ELIF** statement again contains a logical expression and a colon. *ELIF* is short for *else if*. Python evaluate this logical expression, and if TRUE, executes the block of indented code. If it is FALSE, it proceeds to the **ELSE** statement.

The **ELSE** statement is executed if none of the logical expressions are TRUE, so this acts like a "none of the above" option.

### ```while``` loops

There are many times we want to repeat a block of code until some condition is met. 

In these cases, we can use a ```while``` loop:

```python
while logical_expression:
    indented_block_of_code_to_execute_as_long_as_expression_is_true
```

Here's a simple ```while``` loop that loops while the variable x is less than 10:

````python
x = 1

while x < 10:
    print(x)
    x = x + 1

  
````

A cpl things to note.

To enter the ```while``` loop we have to initialize our variable to make sure the ```while``` expression is initially TRUE.

Inside the loop, we have to make some change to the variable, otherwise the its value will remain constant and we will be stuck in the loop. 

Here's another example of a ```while``` loop that determines the minimum launch angle required to throw a ball some specified distance.

Let's say we can throw a ball with an initial velocity of 30 m/s, and we want the ball to go at least 70m. What is the smallest launch angle we can use?

````python
# import math functions, pi and g
from math import cos, sin, pi
from scipy.constants import g

# prompt for initial velocity and distance
v_o = float(input("What is the initial velocity (m/s)?: "))
distance = float(input("How far do you want the ball to go (m)?: "))

# initialize angle in radians
theta = 0.0

# precision of launch angle, in radians
delta_theta = 0.001

# calculate distance based on initial angle
x = 2*v_o**2*sin(theta)*cos(theta)/g

# increment theta until distance is reached
while x <= distance:
    # increment theta by delta_theta
    theta = theta + delta_theta
    # calculate distance using range equation
    x = 2*v_o**2*sin(theta)*cos(theta)/g
    
    
print(f"We need to throw the ball at a minimum angle of {theta*(180./pi):.2f} degrees to go {distance} m.")
````

The block of code inside the *while* loop will execute while the expression *x <= distance* is TRUE. Everytime through the *while* loop we increase the launch angle theta by a small amount, then recalculate how far the ball travels at this angle. The loop continues until x is greater than or equal to our desired distance.

Note, we need to initially assign *x* a value less than *distance* to enter the *while* loop. 

What happens if we specify a distance that is unreachable, no matter what the angle. Rerun the program with a large distance.

The program gets stuck in an infinite loop because the logical expression never is FALSE. To stop the execution of the code, hit the stop button above.

To remedy this, we can use the **break** statement, to break out of the *while* loop if some condition is met.

Since we know the maximum range is achieved when the launch angle is 45 degrees we'd like to break out of the *while* loop if theta exeeds this. Use the **break** statement to do this:

````python
# import math functions, pi and g
from math import cos, sin, pi
from scipy.constants import g

# prompt for initial velocity and distance
v_o = float(input("What is the initial velocity (m/s)?: "))
distance = float(input("What distance do you want the ball to go (m)?: "))

# initialize angle in radians
theta = 0.0

# precision of launch angle, in radians
delta_theta = 0.001

# initialize distance to 0 m in order to enter the while loop
x = 0

# iterate until distance is reached. 
# Break out of the loop if angle exceeds 45 degree
while x <= distance:
    # increment theta by delta_theta
    theta += delta_theta
    # calculate distance using range equation
    x = 2*v_o**2*sin(theta)*cos(theta)/g
    # break out of loop if launch angle exceeds 45 degrees
    if theta > pi/4:
        break
    
# output the minimum angle or the maximam distance.
if theta>pi/4:
    print(f"Sorry, the maximum distance the ball will go is {x:.3f} m.")
else:
    print(f"We need to throw the ball at a minimum angle of {theta*(180./pi):.3f} degrees to go {distance} m.")
````

Within the *while* loop I use an *if* statement to check if the maximum angle has been exceeded. If it is, I use **break** to break out of the loop and return to the main code block. 

Note **break** just breaks out of the current loop. If the loop is nested within another loop, control returns to this loop, not the main code block.

```python
    block of code
```

Notice the ```for``` statement ends with a colon just like ```while``` and ```if``` statements, and the block of code to be executed is indented. The indention is again required and it is what defines the block of code inside the ```for``` loop. 

**\<variable>** can be any variable name. It does not need to be created before use in the ```for``` statement. It is good practice to give it a name that denotes what you are looping over. Often it is just a single character that denotes an index, perhaps i, j, or k.
    
**\<iterable>** is a collection of objects that can be iterated over. Technically an *iterable* is an object that can be passed to the Python function *iter()* to return another object called an *iterator*. The details aren't important but some objects are iterable and some are not. Iterable objects we have encountered so far are *strings*, *lists*, and *tuples*. Scalar objects like integers or floats are not. 


Here is a ```for``` loop over items in a list:

````python
# create a list
l = [2.0, 1, "hello", [1,2,3,4]]

# print each item in the list
for item in l:
    print(item)
````

The ability to loop through different types of objects, not just numbers, makes the *for* loop quite powerful in Python.

### ```range()```

It is often the case we don't want to loop through a collection of things, rather we just want to loop a certain number of times. For example, when integrating a function, we will often divide the x-axis up into N slices, and loop over each slice. 

In these cases it is convenient to use the base function ```range()``` to create a range of values we specify. The syntax of ```range()``` is

```python
range(start_integer, end_integer, integer_increment)
```

All the arguments of ```range()``` must be integers. If you don't specify the increment, the default is an increment of 1:

````python
for i in range(0,10,2):
    print(i)
````

Note that ```range()``` creates a value up to but not including the end_integer, so if we want this to print the value 10 as well, we need to set end_integer to 11.  

Another example that uses an increment of 2:

We can also count down using a negative increment:

````python
for i in range(10, 0, -1):
    print(i)
````

Again note that this counts down to but not including the end_integer. 

Perhaps the most common way of using *range()* is to specify a single argument:

````python
for i in range(10):
    print(i)
````

We see by default, start_integer is set to 0 and increment is set to 1. 

## Nested loops

You should now have a pretty good idea how ```for``` loops work. We will use them a lot this term.  One final thing to note is that we can nest ```for``` loops inside other ```for``` loops. 

Let's look at this and figure out what's happening:

````python
for i in range(5):
    print(i)
    for j in range(3):
        print (i,j)
````

The inner j ```for``` loop is *nested* inside the outer i ```for``` loop.  Once the inner loop cycles 3 times, it breaks back to the outer i loop, which then increments by 1. The inner loop again cycles 3 times and breaks back again to the outer loop. Therefore the ```print()``` statement is executed 5x3=15 times. 

The indentation of the code is again key. What will happen here?

````python
for i in range(5):
    for j in range(3):
    print (i,j)
````

Because the *print()* statement is not indented twice, there is no block of code for the inner loop to execute, and Python complains. So be careful how you indent when working with nested loops.
