**Orbital Mechanics: Determination of Altitude for a Specified Period \(T\)**

The relationship between the orbital period $T$ and altitude $h$ of a satellite in a circular orbit can be derived by considering the balance of gravitational and centripital forces. For a satellite of mass $m$ orbiting a central body of mass $M$ and radius $R$, the gravitational force $F_g$ is given by: 
$$
F_g = \frac{G M m}{r^2} \tag{1}
$$
where $G$ is the Universal gravitational constant and $r = R +h$ represents the orbital radius. For a stable circular orbit, this gravitational attraction attraction must provide the necessary centripital acceleration $a_c = v^2/r$. Equating these two equations gives us
$$
\frac{G M}{r^2} = \frac{v^2}{r} \tag{2}
$$

The orbital velocity is defined by the distance traveled in one period, $v= 2 \pi r/T$. Substituting this into Equation (2) allows us to express the relationship in terms of the period: 
$$
\frac{G M}{r} = \Big ( \frac{2 \pi r}{T} \Big )^2
$$

Rearraing for the orbital radius $r$, we obtain a form consistent with Kepler's Third Law: 
$$
r = \Big ( \frac{GMT^2}{4 \pi^2} \Big)^{1/3}
$$

Lastly, we substitute $r = R + h$, we isolate the required altitude $h$ above the Earth's surface:
$$
h = \Big ( \frac{GMT^2}{4 \pi^2} \Big)^{1/3}  - R
$$

**b)** Write a program that asks the user to enter the desired value of the period *T* and then calculates
and prints out the correct altitude **in meters**. 

Use your program to calculate the altitudes of satellites that orbit the Earth once a day (so-called “geosynchronous” orbit), once every 90 minutes, and once every 45 minutes. 

Report the calculated altitudes in the table below in units of meters using scientific notation to two decimal places. 

```python
# code for part b

from math import pi

# Constants
G = 6.67E-11 
M = 5.97E24
R = 6371E3

# Periods (s)
T_day = 1 *24 * 60 * 60
T_90 = 90 *60 
T_45 = 45 * 60

# User input

T = float(input("Please enter the period the Satellite takes to orbit the Earth in seconds: "))

# Calculate altitude h

def h(T):
    return (G * M * T**2/ (4 * pi**2))**(1/3) - R

# Print output

print(f"For a user input T: {T:0.2f} s, gives us a height of: {h(T):0.2f} m.")
print(f"For a whole day, the height is {h(T_day):0.2f} m.")
print(f"For 90 minutes, the height is: {h(T_90):0.2f} m.")
print(f"For 45 minutes, the height is: {h(T_45):0.2} m.")
```

Catalan Numbers $C_n$

$$
C_0 = 1,  \ \ C_{n+1} = \frac{4n+2}{n+2} C_n
$$

```python
#code 
# Prints in increasing order all Catalan numbers less than or equal to one billion

billion = int(1E9)

def Catalan(C_n, n):
    return (4 * n +2)/(n + 2) * C_n

C = 1

for i in range(billion):
    print(f"C{i} = {C:0.0f}")
    C = Catalan(C, i) 
```

### Exercise 2.9 - Semi-empirical Mass Formula

In nuclear physics, the semi-empirical mass formula is a formula for calculating the approximate nuclear binding energy _B_ of an atomic nucleus with atomic number _Z_ and mass number _A_:

$B = a_{1}A - a_{2}A^{2/3} - a_{3}\frac{Z^2}{A^{1/3}} - a_{4}\frac{(A-2Z)^{2}}{A} + \frac{a_{5}}{A^{1/2}}, $

where, in units of millions of electron volts, the constants are $a_{1}$ = 15.8, $a_{2}$ = 18.3, $a_{3}$ = 0.714, $a_{4}$ = 23.2, and

\begin{equation*}
a_{5} = \begin{cases}
            0 \quad &\text{if A is odd,} \\
            12.0 \quad &\text{if A and Z are both even,} \\
            -12.0 \quad &\text{if A is even and Z is odd.} \\
        \end{cases}
\end{equation*}

**a)**
Write a program that takes as its input the values of _A_ and _Z_, and prints out the binding energy per nucleon, $B/A$, for the corresponding atom. 

To test your program, have it print out the binding energy per nucleon of an atom with _A_ = 58 and _Z_ = 28 in units of millions of electron volts (MeV). Report to two decimal places. (Hint: The correct answer is around 8.6 MeV per nucleon.)

```python
# 2.10a


# Constants

a_1 = 15.8
a_2 = 18.3
a_3 = 0.714
a_4 = 23.2
a_5 = 0.0
# User Input

A = float(input("Please enter the mass number: "))
Z = float(input("Please enter the atomic number: "))

if A%2 == 1.0: 
    a_5 = 0.0
elif (A%2) == 0.0 and (Z%2) == 0.0:
    a_5 = 12.0
else:
    a_5 = -12.0
    
# Equation

B = a_1 * A - a_2 * A**(2/3) - a_3 * (Z**2)/(A**(1/3)) -(a_4)* (A- 2 * Z)**2/A + (a_5)/(A**(1/2))

# Print

print(f"The Binding Energy is: {B/A} MeV.")
```
