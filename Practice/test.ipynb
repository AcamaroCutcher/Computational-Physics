## 1. Variables & Data Types

````python
# Basic types
x = 42          # int
y = 3.14        # float
name = 'Alice'  # str
flag = True     # bool
nothing = None  # NoneType

# Type conversion
print(int('7'), float(3), str(99), bool(0))

# Multiple assignment
a, b, c = 1, 2, 3
p = q = 0

# f-strings
print(f'{name} is {x} years old and pi ≈ {y:.4f}')
````

## 2. Collections

````python
# List — ordered, mutable
nums = [1, 2, 3, 4, 5]
nums.append(6)
nums.pop()              # removes last
print(nums[0], nums[-1], nums[1:3])  # indexing & slicing

# Tuple — ordered, immutable
coords = (10.0, -3.5)
cx, cy = coords         # unpacking

# Dict — key-value pairs
d = {'a': 1, 'b': 2}
d['c'] = 3
print(d.get('z', 0))   # safe lookup with default
for k, v in d.items():
    print(k, v)

# Set — unique, unordered
s = {1, 2, 2, 3}
s.add(4)
print(s)
````

## 3. Control Flow

````python
# if / elif / else
score = 85
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
else:
    grade = 'C'
print(grade)

# for loop
for i in range(5):
    print(i, end=' ')
print()

for i, val in enumerate(['a', 'b', 'c']):
    print(i, val)

# while loop
n = 10
while n > 0:
    n -= 3
print(n)

# break / continue
for i in range(10):
    if i == 3: continue
    if i == 6: break
    print(i, end=' ')
print()
````

## 4. Functions

````python
# Basic function
def add(a, b):
    return a + b

# Default & keyword args
def greet(name, greeting='Hello'):
    return f'{greeting}, {name}!'

# *args and **kwargs
def total(*args):
    return sum(args)

def info(**kwargs):
    for k, v in kwargs.items():
        print(f'{k}: {v}')

# Lambda
square = lambda x: x ** 2

# Recursion
def factorial(n):
    return 1 if n <= 1 else n * factorial(n - 1)

print(add(3, 4))
print(greet('Bob'))
print(total(1, 2, 3, 4))
info(lang='Python', version=3.12)
print(square(5))
print(factorial(6))
````

## 5. List Comprehensions & Generators

````python
# List comprehension
squares = [x**2 for x in range(10)]
evens   = [x for x in range(20) if x % 2 == 0]

# Dict comprehension
sq_map = {x: x**2 for x in range(5)}

# Generator expression (lazy — doesn't build list in memory)
gen = (x**2 for x in range(1_000_000))
print(next(gen), next(gen))

# map / filter
doubled = list(map(lambda x: x*2, range(5)))
odds    = list(filter(lambda x: x % 2, range(10)))

print(squares)
print(evens)
print(doubled)
print(odds)
````

## 6. Classes & OOP

````python
class Animal:
    species = 'Unknown'          # class variable

    def __init__(self, name, sound):
        self.name  = name        # instance variable
        self.sound = sound

    def speak(self):
        return f'{self.name} says {self.sound}'

    def __repr__(self):
        return f'Animal({self.name!r})'


class Dog(Animal):               # inheritance
    species = 'Canis lupus'

    def __init__(self, name):
        super().__init__(name, 'Woof')

    def fetch(self, item):
        return f'{self.name} fetches the {item}!'


d = Dog('Rex')
print(d.speak())
print(d.fetch('ball'))
print(repr(d))
print(isinstance(d, Animal))
````

## 7. Error Handling

````python
# try / except / else / finally
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f'Error: {e}')
except (TypeError, ValueError):
    print('Type or value error')
else:
    print('No error')
finally:
    print('Always runs')

# Raising exceptions
def sqrt_safe(x):
    if x < 0:
        raise ValueError(f'Cannot take sqrt of {x}')
    return x ** 0.5

print(sqrt_safe(9))

# Context manager
# with open('data.txt', 'w') as f:
#     f.write('hello')
````

## 8. NumPy Essentials

````python
import numpy as np


# Array creation
a = np.array([1, 2, 3, 4])
z = np.zeros((3, 3))
I = np.eye(3)
r = np.linspace(0, 1, 5)         # 5 evenly spaced values in [0,1]
rng  = np.random.default_rng(42)
rand = rng.random((2, 3))

# Vectorized math — no Python loops needed
print(a * 2)
print(np.sin(np.pi / 2))
print(a.dot(a))                  # dot product == 30

# 2D indexing & slicing
m = np.arange(9).reshape(3, 3)
print(m[1, :])                   # second row
print(m[:, 2])                   # third column
print(m[m > 4])                  # boolean mask
````

## 9. Matplotlib Essentials

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 300)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(x, np.sin(x), label='sin')
axes[0].plot(x, np.cos(x), label='cos')
axes[0].set_title('Trig Functions')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].legend()

rng = np.random.default_rng(0)
axes[1].scatter(rng.standard_normal(100), rng.standard_normal(100), alpha=0.5)
axes[1].set_title('Scatter Plot')

plt.tight_layout()
plt.show()
````
